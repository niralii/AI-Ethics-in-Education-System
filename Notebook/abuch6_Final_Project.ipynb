## 3) Select a pre-processing bias mitigation algorithm to transform the original dataset (e.g. Reweighting, Disparate Impact Remover, etc.) as a function of one of your dependent variables

```python
# bias mitigation algorithm - DisparateImpactRemover
# dependent variables - math_result
```

```python
data.loc[data.gender == 'male', 'gender'] = 1
data.loc[data.gender == 'female', 'gender'] = 0

# data.loc[data.reading_result == 'Pass', 'reading_result'] = 1
# data.loc[data.reading_result == 'Fail', 'reading_result'] = 0

# data.loc[data.writing_result == 'Pass', 'writing_result'] = 1
# data.loc[data.writing_result == 'Fail', 'writing_result'] = 0

data.loc[data.math_result == 'Pass', 'math_result'] = 1
data.loc[data.math_result == 'Fail', 'math_result'] = 0

data.loc[data.reading_result == 'Pass', 'reading_result'] = 1
data.loc[data.reading_result == 'Fail', 'reading_result'] = 0

data.loc[data.writing_result == 'Pass', 'writing_result'] = 1
data.loc[data.writing_result == 'Fail', 'writing_result'] = 0

data.loc[data.overall_result == 'Pass', 'overall_result'] = 1
data.loc[data.overall_result == 'Fail', 'overall_result'] = 0

data.loc[data.test_preparation_course == 'completed', 'test_preparation_course'] = 1
data.loc[data.test_preparation_course == 'none', 'test_preparation_course'] = 0

# data.loc[data.overall_result == 'Pass', 'overall_result'] = 1
# data.loc[data.overall_result == 'Fail', 'overall_result'] = 0
```

```python
data.head()
```

```python

binaryLabelDataset = BinaryLabelDataset(
    favorable_label=1,
    unfavorable_label=0,
    df=data[['gender','race','overall_result','test_preparation_course','math_score','reading_score','writing_score','math_result','reading_result','writing_result']],
    label_names=['math_result'],
    protected_attribute_names=['gender'],
)
```

```python
di = DisparateImpactRemover(repair_level=1, sensitive_attribute='gender')
transformed_data = di.fit_transform(binaryLabelDataset)
transformed = transformed_data.convert_to_dataframe()[0]
transformed
```

## 1) Randomly split your original dataset into training and testing datasets

Here, from the original dataset, we choose 'overall_result' as the outcome / dependent variable.

```python
features = ['gender','race','test_preparation_course','math_score','reading_score','writing_score','math_result','reading_result','writing_result']

X_original = data.loc[:, features]
X_original = X_original.convert_dtypes()
y_original = data.loc[:, ['overall_result']]
y_original = y_original.convert_dtypes()
X_original_train, X_original_test, y_original_train, y_original_test = train_test_split(X_original, y_original, random_state=randrange(100), train_size = 0.65, shuffle=True)
```

## 2) Randomly split your transformed dataset into training and testing datasets

```python
X_transformed = transformed.loc[:, features]
X_transformed = X_transformed.convert_dtypes()
y_transformed = transformed.loc[:, ['overall_result']]
y_transformed = y_transformed.convert_dtypes()
X_transformed_train, X_transformed_test, y_transformed_train, y_transformed_test = train_test_split(X_transformed, y_transformed, random_state=randrange(100), train_size = 0.65)
```

## 3) Train a classifier using the original training dataset from Step 4.1; select one of your dependent variables as the output label to train your classifier.

```python
logisticRegr = LogisticRegression(solver='liblinear')

logisticRegr.fit(X_original_train, y_original_train['overall_result'])

predictions_original = logisticRegr.predict(X_original_test)

score = logisticRegr.score(X_original_test, y_original_test)
print(f'The accuracy of the model on the original dataset is {score}')
```

## 4) Train a classifier using the transformed training dataset from Step 4.2; select one of your dependent variables as the output label to train your classifier.

```python
logisticRegr = LogisticRegression(solver='liblinear')

logisticRegr.fit(X_transformed_train, y_transformed_train['overall_result'])

predictions_transformed = logisticRegr.predict(X_transformed_test)

score = logisticRegr.score(X_transformed_test, y_transformed_test)
print(f'The accuracy of the model on the transformed dataset is {score}')
```
